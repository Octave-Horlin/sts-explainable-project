# Analysis

Load metrics/ablations and inspect comparisons.

```python
import json
from pathlib import Path
import pandas as pd

base = Path("..").resolve() / "metrics"
stsb_base = json.loads((base / "stsb_baseline.json").read_text()) if (base / "stsb_baseline.json").exists() else {}
stsb_adv = json.loads((base / "stsb_advanced.json").read_text()) if (base / "stsb_advanced.json").exists() else {}
sick_adv = json.loads((base / "sick_advanced.json").read_text()) if (base / "sick_advanced.json").exists() else {}
abl_stsb = json.loads((base / "ablations_stsb.json").read_text()) if (base / "ablations_stsb.json").exists() else {}
abl_sick = json.loads((base / "ablations_sick.json").read_text()) if (base / "ablations_sick.json").exists() else {}

stsb_base, stsb_adv, sick_adv, abl_stsb, abl_sick
```

```python
rows = []
if stsb_base:
    rows.append({"dataset": "stsb", "model": "baseline", **stsb_base})
if stsb_adv:
    rows.append({"dataset": "stsb", "model": "tuned", **{k:v for k,v in stsb_adv.items() if k!='split'}})
if sick_adv:
    rows.append({"dataset": "sick", "model": "tuned", **{k:v for k,v in sick_adv.items() if k!='split'}})
pd.DataFrame(rows)
```
