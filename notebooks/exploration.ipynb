Quick checks on STS-B and SICK processed datasets. Adjust paths if running outside repo root.

```python
import pandas as pd
from pathlib import Path

base = Path("..").resolve() / "data" / "processed"
stsb_train = pd.read_csv(base / "stsb_train.csv")
sick_train = pd.read_csv(base / "sick_train.csv")

stsb_train.describe()
```
